```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
genre_load = "Resources/pre-final-88-genre.csv"

# Read School and Student Data File and store into Pandas DataFrames
genre_data = pd.read_csv(genre_load)
```

```python
df = genre_data.drop(["Unnamed: 0","track", "album", "artist", "popularity", "featured", "release_date", "track_id", "time_signature"], axis=1)
```

```python
df
```

```python
df['genre']
```

```python
df_drop =  df.drop(df[df['genre'] == '[]'].index)
df_drop
```

```python
def extract_genre(x):
    if isinstance(x, list):
        return x[0] if x else None
    elif isinstance(x, str):
        # Remove brackets from the string and return the first genre tag
        return x.strip('[]').split(',')[0] if x else None
    else:
        return None
```

```python
df_filtered = df_drop['primary_genre'] = df_drop['genre'].apply(extract_genre)
df_filtered
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
genres_df = [rap_lyrics_df, rb_lyrics_df, pop_lyrics_df, rock_lyrics_df]
colormaps = ["Purples", "Blues", "Reds", "Greens"]
```

```python
for i in range(len(genres_df)):
  stopwords = set(STOPWORDS)
  wordcloud = WordCloud(stopwords=stopwords, background_color=None,colormap=colormaps[i], max_words=800).generate(" ".join(str(x) for x in (genres_df[i]["filtered_lyrics"])))
  plt.figure(figsize=(12,12))
  plt.imshow(wordcloud, interpolation="bilinear")
  plt.show()
```
